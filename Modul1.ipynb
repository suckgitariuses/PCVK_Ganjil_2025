**Penggunaan Github dan Google Colaboratory Membaca dan Menampilkan Citra**

**Step 1**

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

**Step 2**

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://i.pinimg.com/736x/ba/5f/c2/ba5fc2d25b2d1a7ce02852d95330bd24.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                           # read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  # resize image to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # convert color to RGB
    final_frame = cv.hconcat([image, image_2])       # concatenate image
    cv2_imshow(final_frame)                          # show image
    print('\n')

```

**Step 3**

Melihat ukuran File Image


```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

**Step 4**

mengakses pixel dengan memberi garis diagonal menyilang

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal di tengah image
for y in range(lebar):
    image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
urls = ["https://i.pinimg.com/736x/2f/6b/78/2f6b784dace8e109293956006910884f.jpg"]

# baca dan tampilkan image
for url in urls:
    image = io.imread(url)                           # read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  # resize image to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # convert color to RGB

    # ambil ukuran citra
    tinggi, lebar = image_2.shape[0], image_2.shape[1]

    # ---------- Gambar dengan garis vertikal ----------
    img_vertikal = image_2.copy()
    for y in range(tinggi):
        img_vertikal[y, int(lebar/2)] = [255, 255, 255]

    # ---------- Gambar dengan garis diagonal menyilang ----------
    img_diagonal = image_2.copy()
    # diagonal kiri atas -> kanan bawah
    for i in range(min(tinggi, lebar)):
        img_diagonal[i, i] = [255, 255, 255]
    # diagonal kiri bawah -> kanan atas
    for i in range(min(tinggi, lebar)):
        img_diagonal[tinggi - i - 1, i] = [255, 255, 255]

    # tampilkan hasil
    cv2_imshow(img_vertikal)
    print("Gambar dengan garis vertikal\n")

    cv2_imshow(img_diagonal)
    print("Gambar dengan garis diagonal menyilang\n")
```

3. Buat garis horizontal berwarna putih dibagian tengah gambar dengan panjang tertentu

```python
urls = ["https://i.pinimg.com/736x/2f/6b/78/2f6b784dace8e109293956006910884f.jpg"]

for url in urls:
    image = io.imread(url)
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)   # resize
    image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB) # convert ke RGB

    # ambil ukuran
    h, w = image_rgb.shape[:2]

    # koordinat garis (horizontal di tengah, panjang setengah lebar gambar)
    start_point = (int(w/4), int(h/2))   # titik awal
    end_point   = (int(3*w/4), int(h/2)) # titik akhir

    # gambar garis putih tebal 2px
    img_line = image_rgb.copy()
    cv.line(img_line, start_point, end_point, (255,255,255), 2)

    # tampilkan
    cv2_imshow(img_line)
    print("Gambar dengan garis horizontal putih di tengah\n")
```

4. buat kotak menggunakan kumpulan pixel warna putih di sembarang tembat dalam gambar

```python
urls = ["https://i.pinimg.com/736x/2f/6b/78/2f6b784dace8e109293956006910884f.jpg"]

for url in urls:
    image = io.imread(url)
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)   # resize
    image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB) # convert ke RGB

    # buat copy gambar
    img_box = image_rgb.copy()

    # koordinat kotak (x1,y1) ke (x2,y2)
    start_point = (100, 80)     # pojok kiri atas
    end_point   = (200, 180)   # pojok kanan bawah

    # gambar kotak putih tebal 2px
    cv.rectangle(img_box, start_point, end_point, (255,255,255), -1)

    # tampilkan
    cv2_imshow(img_box)
    print("Kotak putih berhasil dibuat\n")
```
