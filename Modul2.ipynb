**2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!**

**Jawab:**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# OpenCV baca BGR
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/kitten01.jpg')

# Red-Blue
# buat salinan gambar, atur Green menjadi 0
img_RB = img.copy()
img_RB[:, :, 1] = 0 # atur Green (index 1) ke 0

# Green-Blue (tanpa Red)
# Buat salinan gambar, atur Red menjadi 0
img_GB = img.copy()
img_GB[:, :, 2] = 0  # atur Red (index 2) ke 0

# Tampilkan
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Red-Blue (no Green)')
plt.imshow(cv2.cvtColor(img_RB, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Green-Blue (no Red)')
plt.imshow(cv2.cvtColor(img_GB, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

**5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Buat canvas kosong (hitam), ukuran 400x400 px, 3 channel warna
img = np.zeros((400, 400, 3), dtype=np.uint8)

# Bikin 5 kotak random
for i in range(5):
    # Koordinat random untuk pojok kiri atas dan kanan bawah
    x1, y1 = np.random.randint(0, 300, size=2)
    x2, y2 = np.random.randint(x1+20, 400, size=1)[0], np.random.randint(y1+20, 400, size=1)[0]

    # Warna random (BGR format OpenCV)
    color = tuple(np.random.randint(0, 256, size=3).tolist())

    # Gambar kotak
    cv2.rectangle(img, (x1, y1), (x2, y2), color, -1)  # -1 biar filled

# Tampilkan hasil
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("5 Kotak Random dengan Warna & Ukuran Berbeda")
plt.axis("off")
plt.show()
```

**6. Tampilkan image dengan posisi terbalik!**

```python
import cv2
import matplotlib.pyplot as plt

# Baca gambar
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/kitten01.jpg')

# Flip vertical (atas-bawah)
flip_vertical = cv2.flip(img, 0)

# Flip horizontal (kiri-kanan)
flip_horizontal = cv2.flip(img, 1)

# Flip keduanya (180 derajat)
flip_both = cv2.flip(img, -1)

# Tampilkan hasil
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,4,2)
plt.title("Vertical Flip")
plt.imshow(cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,4,3)
plt.title("Horizontal Flip")
plt.imshow(cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,4,4)
plt.title("Both (180)")
plt.imshow(cv2.cvtColor(flip_both, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

**7. Buat rectangle dan circle pada bagian wajah dari image foto anda saat
beraktifitas (bukan pasfoto).**

**8. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB
dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!**

```python
import cv2
import matplotlib.pyplot as plt

# Baca citra
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/ganteng.jpg')

# Pastikan terbaca
if img is None:
    raise FileNotFoundError("Gambar tidak ditemukan!")

# Ambil channel B (index 0 pada BGR)
blue_channel = img[:,:,0]

# Buat salinan grayscale dari channel B
blue_img = cv2.merge([blue_channel, blue_channel, blue_channel])

# Dapatkan ukuran citra
h, w = blue_channel.shape

# Gambar rectangle di pojok kiri bawah (misal ukuran 100x100 px)
cv2.rectangle(blue_img, (0, h-100), (100, h), (255, 0, 0), 3)

# Tampilkan
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(blue_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Rectangle pada Channel B (pojok kiri bawah)")
plt.show()
```

**9. Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran
font, dan warna font yang sesuai keinginan anda.**

```python
import cv2
import matplotlib.pyplot as plt

# Baca citra (contoh kitten01.jpg, ganti sesuai file lain: lena, mandrill, dll)
filename = "ganteng.jpg"
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/ganteng.jpg')

if img is None:
    raise FileNotFoundError("Gambar tidak ditemukan!")

# Ambil channel B
blue_channel = img[:,:,0]
blue_img = cv2.merge([blue_channel, blue_channel, blue_channel])

# Ukuran citra
h, w = blue_channel.shape

# Buat rectangle di pojok kiri bawah
cv2.rectangle(blue_img, (0, h-100), (120, h), (255, 0, 0), 3)

# Tambahkan tulisan nama file
cv2.putText(
    blue_img,                   # citra target
    filename,                   # teks yang ditampilkan
    (10, h-110),                # posisi (x,y) teks
    cv2.FONT_HERSHEY_SIMPLEX,   # jenis font
    0.8,                        # ukuran font
    (0, 255, 0),                # warna (hijau)
    2,                          # ketebalan
    cv2.LINE_AA                 # anti alias
)

# Tampilkan hasil
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(blue_img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Rectangle + Ganteng")
plt.show()
```
